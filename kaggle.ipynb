```python
d=cr['date']
p=cr['price']
pc=cr['percentChange']
c=cr['change']
```

```python
plt.plot(p,c)
```

```python
plt.plot(d,color='r')
```

```python
plt.plot(p,pc,color='y')
plt.title('PRICE VS PERCENTCHANGE')
plt.xlabel('PRICE')
plt.ylabel('PERCENT CHANGE')
```

```python
plt.hist(cr['price'],bins=40,color='b')
plt.title(" PRICE GRAPH")
```

```python
plt.bar(p,pc,color='pink')
plt.xlabel('PRICE')
plt.ylabel('PERCENT CHANGE')
```

```python
plt.bar(d,p,color='orange')
plt.xlabel('DATE')
plt.ylabel('PRICE')
```

```python
plt.scatter(p,pc,marker='.',color='g')
```

```python
plt.scatter(p,pc,marker='*',c='g',s=50)
plt.scatter(c,pc,marker='.',c='b',s=20)
```

```python
plt.subplot(1,2,1)
plt.scatter(p,pc,marker='*',c='black',s=100)
plt.title('SUBPLOT')
plt.xlabel('PRICE')
plt.ylabel('PERCENT CHANGE')


plt.subplot(1,2,2)
plt.scatter(c,p,marker='.',c='red',s=100)
plt.title('SUBPLOT')
plt.xlabel('CHANGE')
plt.ylabel(' PRICE')
```
