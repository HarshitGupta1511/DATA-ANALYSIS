```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cr=pd.read_csv('crude-oil-price.csv')
```

```python
d=cr['date']
p=cr['price']
pc=cr['percentChange']
c=cr['change']
```

```python
plt.plot(d,p)
```

```python
plt.plot(c,color='pink')
```

```python
plt.plot(d,pc,color='y')
plt.title('DATE VS PERCENTCHANGE')
plt.xlabel('DATE')
plt.ylabel('PERCENT CHANGE')
```

```python
plt.hist(cr['price'],bins=40,color='b')
plt.title(" PRICE GRAPH")
```

```python
plt.scatter(p,pc,marker='.',color='g')
```

```python
import seaborn as sns

#we need to declare and pass data while using seaborn , give data=whatever value you have assigned while reading , in above case its cr
```

```python
sns.lineplot(x="date", y="price",data=cr)
```

```python
palette = sns.color_palette()
```

```python
sns.palplot(palette)
```

```python
def plot():
    sns.lineplot(x="date", y="change", data=cr)

sns.set_palette('vlag')
plt.subplot(211)

sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

```python
sns.relplot(x='change', y='date', data=cr)

plt.show()
```

```python
sns.scatterplot(x='price', y='change', data=cr,color='orange')
plt.show()
```

```python
sns.stripplot(x='date', y='price', data=cr)
plt.show()
```
